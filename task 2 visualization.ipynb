```python
plt.figure(figsize=(14, 5))

# Subplot 1
plt.subplot(1, 3, 1)
ax1 = sns.countplot(x='comments_on/off', data=df)
for p in ax1.patches:
    ax1.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Subplot 2
plt.subplot(1, 3, 2)
ax2 = sns.countplot(x='ratings_on/off', data=df)
for p in ax2.patches:
    ax2.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Subplot 3
plt.subplot(1, 3, 3)
ax3 = sns.countplot(x='removed', data=df)
for p in ax3.patches:
    ax3.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python
# these bar plots depicts the no.of video in which comments are disabled,ratings are disabled and the video which are removed
```

```python
plt.figure(figsize=(12, 15))

plt.subplot(3, 3, 1)
sns.boxplot(x='comments_disabled', y='views', data=df)

plt.subplot(3, 3, 2)
sns.boxplot(x='ratings_disabled', y='views', data=df)

plt.subplot(3, 3, 3)
sns.boxplot(x='video_error_or_removed', y='views', data=df)

plt.subplot(3, 3, 4)
sns.boxplot(x='comments_disabled', y='likes', data=df)

plt.subplot(3, 3, 5)
sns.boxplot(x='ratings_disabled', y='likes', data=df)

plt.subplot(3, 3, 6)
sns.boxplot(x='video_error_or_removed', y='likes', data=df)


plt.show()
```

```python
# these plots show that the videos in which either the comments or ratings are removed or the video is removed mostly have least views or likes thus those are low quality videos or the content is poor
```

```python
corr_matrix = df[['views', 'likes', 'dislikes', 'comment_count']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# here we are getting the correlating coefficients between 'views', 'likes', 'dislikes', 'comment_count'  and we have converted the values of correlating coefficients into a heat format
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already loaded your DataFrame df

# Define the columns for analysis
columns_of_interest = ['views', 'likes']

# Create a pair plot to visualize outliers
sns.pairplot(df[columns_of_interest], height=4)

# Identify outliers using the IQR method
Q1 = df[columns_of_interest].quantile(0.25)
Q3 = df[columns_of_interest].quantile(0.75)
IQR = Q3 - Q1

# Define a function to identify outliers
def identify_outliers(column):
    return (df[column] < (Q1[column] - 1.5 * IQR[column])) | (df[column] > (Q3[column] + 1.5 * IQR[column]))

# Identify outliers for 'views' and 'likes'
outliers_views = identify_outliers('views')
outliers_likes = identify_outliers('likes')

# Highlight outliers in red on the pair plot
for ax in plt.gcf().axes:
    ax.scatter(df.loc[outliers_views, 'views'], df.loc[outliers_views, 'likes'], color='red', marker='x', s=100)
    ax.scatter(df.loc[outliers_likes, 'views'], df.loc[outliers_likes, 'likes'], color='red', marker='x', s=100)

plt.show()
```

```python
# these plots mark out all the outliers --the videos which might be having very high no.of views but has less likes ,or simply the videos with abnormally high views
```
